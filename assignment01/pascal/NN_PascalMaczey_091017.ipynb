```python
import numpy as np
import sympy as sp
sp.init_printing()
```

Import of libraries and initializing for printing

```python
a, v = sp.symbols("a, v")
```

Creating symbols for functions

## 1.1

```python
f1 = 1/(1+sp.exp(-a*v))
f1
```

Creating the function

```python
f2 = sp.diff(f1, v)
f2
```

Calculating the derivation

```python
f3 = f2 - a * f1 * (1 -f1)
f3 = sp.simplify(f3)
f3
```

To verify both derivations (given from exercise and the caculated one) are equal, the difference is caculated. The result is zero. That means both derivations are equal.

```python
f2.subs(v,0)
```

The Value of the derivation at the origin is $\frac{a}{4}$

## 1.2

```python
g1 = (1 - sp.exp(-a*v)) / (1 + sp.exp(-a*v))
g1 = sp.simplify(g1)
g1
```

Creating the function

```python
g2 = sp.diff(g1, v)
g3 = sp.simplify(g2 - (a/2)*(1 - g1**2))
g3
```

Calculating the derivation and verifying the calculated derivation and the given derivation are equal

```python
g2.subs(v, 0)
```

$\frac{a}{2}$ is the value of the derivation at the origin
