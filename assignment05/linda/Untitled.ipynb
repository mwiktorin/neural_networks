```python
import numpy as np
import sympy as sp
sp.init_printing()
```

```python
w_1,w_2 = sp.symbols("w_1,w_2 ")

r_xd=100;
r_x=100;
o=100;

def function(x,y):
    return (1/2)*o-r_xd*x+(1/2)*r_x*y
```

```python
sp.plotting.plot3d(function(w_1,w_2))
```

```python
r_xd=100;
r_x=1000;
o=100;

sp.plotting.plot3d(function(w_1,w_2))
```

```python
r_xd=100;
r_x=-1000;
o=100;

sp.plotting.plot3d(function(w_1,w_2))
```

$r_x$ rotates the hyperplane
